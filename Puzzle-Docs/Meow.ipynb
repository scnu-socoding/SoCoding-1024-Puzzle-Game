```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('./meow_small.png', cv.IMREAD_UNCHANGED)
plt.imshow(img)
```

```python
def arnold(img):
    r, c, _ = img.shape
    p = np.zeros(img.shape, np.uint8)
    a = -1
    b = -1
    for i in range(r):
        for j in range(c):
            x = (i + b * j) % r
            y = (a * i + (a * b + 1) * j) % c
            p[x, y, :] = img[i, j, :]
    return p


def dearnold(img):
    r, c, _ = img.shape
    p = np.zeros(img.shape, np.uint8)
    a = -1
    b = -1
    for i in range(r):
        for j in range(c):
            x = ((a * b + 1) * i - b * j) % r
            y = (-a * i + j) % c
            p[x, y, :] = img[i, j, :]
    return p
```

```python
img2 = arnold(img)
plt.imshow(img2)
cv.imwrite('meow-tr.png', img2)
```

```python
img3 = dearnold(img2)
plt.imshow(img3)
```

```python
for i in range(1, 90):
    plt.subplot(9, 10, i)
    plt.imshow(img)
    img = arnold(img)
```
